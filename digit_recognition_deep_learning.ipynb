```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('digit_data/train.csv')
train.head()
```

```python
test = pd.read_csv('digit_data/test.csv')
test.head()
```

```python
y_train = train['label']
train = train.drop('label', axis=1)

train = np.array(train)
X_train = train.reshape(train.shape[0], 28, 28)
X_test = np.array(test)

X_train = X_train.reshape(X_train.shape[0], 28, 28,1)
X_test = X_test.reshape(X_test.shape[0], 28, 28,1)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
```

```python
X_train.min(), X_train.max()
```

```python
X_test.min(), X_train.max()
```

```python
X_train = X_train.astype('float64')/255
X_test = X_test.astype('float64')/255
```

```python
assert ((X_train.min(),X_train.max()) == (0.0,1.0))
assert ((X_test.min(),X_test.max()) == (0.0,1.0))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(28,28,1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(64, (5, 5), padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(10) #10 is the number of classes
])

model.compile(optimizer='adam',
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True,reduction='auto'),
        metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=10)
```

```python
predictions = np.argmax(model.predict(X_test), axis=1)
```

```python
submission = pd.read_csv('digit_data/sample_submission.csv')
print(submission.shape)
submission.head()
```

```python
submission['Label']  = predictions
submission.head()
```
